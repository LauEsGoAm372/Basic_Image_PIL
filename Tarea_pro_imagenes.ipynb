#1. Abrir imagen y describir:
* Tipo de imagen
* Canales
* Resolución
* Cuantización (número de bits)

```python
from PIL import Image
import matplotlib.pyplot as plt
import requests
from io import BytesIO
```

```python
#Se obtiene la imagen de internet para no estar cargando la imagen constantemente
ruta = "https://thesweetmolcajete.com/wp-content/uploads/2020/06/WPRM-Receta-de-Pastel-de-Tres-Leches.jpg"
response = requests.get(ruta)
imagenB = BytesIO(response.content)
```

```python
#Se abre la imagen
imagensita = Image.open(imagenB)
```

```python
#Se muestra la imagen abierta
plt.imshow(imagensita)
plt.title("Imagen leida")
```

```python
#Para saber que tipo o formato de imagen es se usa format
print("Tipo de imagen: ",imagensita.format)
#Para obtener los canales se usa getbands
print("Canales: ",imagensita.getbands())
#Para obtener la resolución o tamaño de la imagen se usa size
print("Resolución: ",imagensita.size)
#Para saber su cuantización (CUANTOS BITS REPRESENTA) se usa bits
print("Cuantización: ", imagensita.bits)
```

#2. Extraer un canal

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#imagen inicial
axs[0,0].imshow(imagensita)
axs[0,0].set_title("original")

#Obtener el canal rojo
imagensita_roja = imagensita.getchannel("R")
axs[0,1].imshow(imagensita_roja)
axs[0,1].set_title("Canal Rojo")

#Obtener el canal azul
imagensita_verde = imagensita.getchannel("G")
axs[1,0].imshow(imagensita_verde)
axs[1,0].set_title("Canal Verde")

#Obtener el canal azul
imagensita_azul = imagensita.getchannel("B")
axs[1,1].imshow(imagensita_azul)
axs[1,1].set_title("Canal Azul")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

#imagen inicial
axs[0,0].imshow(imagensita, cmap = "gray")
axs[0,0].set_title("original")

#Obtener el canal rojo
imagensita_roja = imagensita.getchannel("R")
axs[0,1].imshow(imagensita_roja, cmap = "gray")
axs[0,1].set_title("Canal Rojo")

#Obtener el canal azul
imagensita_verde = imagensita.getchannel("G")
axs[1,0].imshow(imagensita_verde, cmap = "gray")
axs[1,0].set_title("Canal Verde")

#Obtener el canal azul
imagensita_azul = imagensita.getchannel("B")
axs[1,1].imshow(imagensita_azul,cmap = "gray")
axs[1,1].set_title("Canal Azul")
```
